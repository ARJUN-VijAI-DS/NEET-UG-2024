<font size = 5.5> **Objective and Explanation of Features of Data Set**

<font size = 4.5>The objective of this data set is to provide detailed information and evidence to support the claim that there was a significant leak of examination papers in the year 2024. By examining various metrics across different regions and exam centers, the data aims to highlight discrepancies and anomalies that indicate malpractice, specifically paper leaks. The key elements include:
    <br><br>**Identification of Paper Leaks:** The data categorizes instances of paper leaks across different regions and exam centers.
    <br><br>**Quantification of Cases:** The number of paper leak cases is recorded, showing regions and centers with higher occurrences.
    <br><br>**Impact Analysis:** The impact of these leaks on examinations, such as the cancellation of exams and changes in average marks, topper marks, and the number of successful candidates.
    <br><br>**Actions Taken:** Documentation of actions taken in response to these leaks, such as the cancellation of exams, to demonstrate the severity and response to the issue.
    <br><br>**Performance Metrics:** Analysis of average marks, topper marks, number of toppers, number of successful candidates, and the number of perfect scores to identify irregularities that could indicate the influence of leaked papers.
    <br><br>By aggregating and analyzing this data, we aim to present a comprehensive view that supports the assertion of paper leaks, demonstrating their occurrence, impact, and the measures taken in response.

<font size = 4.5> **Description of All Columns**
    <br><br>**Year:** Represents the calendar year in which the data was collected, specifically focusing on the year 2024.
    <br><br>**Region:** Indicates the geographical area where the examination centers are located. The regions included in the data are North, South, East, and West.<br><br>
    **Exam Center:** Specifies the individual examination center within each region. Each center is denoted by a letter (A, B, C, D).
    <br><br>**Type of Scam:** Describes the type of irregularity or malpractice reported at the exam center. This column highlights whether there was a 'Paper Leak' or if there were 'None' (no scams reported).
    <br><br>**Number of Cases:** Denotes the number of reported cases of the specified type of scam (e.g., paper leaks) at each exam center.
    <br><br>**Actions Taken:** Lists the measures implemented in response to the reported scams. Actions could include 'Exams Cancelled' or 'None' (indicating no actions were taken).
    <br><br>**Average Marks:** Shows the average score obtained by candidates in each region and exam center.
    <br><br>**Topper Marks:** Displays the highest marks achieved by the top-performing student(s) at each exam center.
    <br><br>**Number of Toppers:** Indicates the number of students who scored the highest marks at each exam center.
    <br><br>**Number of Successful Candidates:** Represents the total number of candidates who passed the examination at each exam center.
    <br><br>**Number of Perfect Scores:** Shows the number of students who achieved a perfect score in the examination at each exam center.
    <br><br>These columns collectively provide a comprehensive overview of the examination process, the occurrence of paper leaks, and their subsequent impact on the performance and outcomes of the candidates.

<font size = 5.5> **Data Visualization** <br><br>Graphical representations make it easier to see patterns, trends, and outliers.

<font size = 5.5>**1. Histograms:** Show the distribution of a single numerical variable.<br><br>**2. Box Plots:** Box plots help identify outliers and the spread of the data.<br><br>**3. Scatter Plot:** Used to visualize the relationship between two numerical variables.

```python
import seaborn as sns             # Visualization Library
import matplotlib.pyplot as plt   # Visualization Library

# Assuming DF_COPY is already defined as the DataFrame
for col in DF_COPY.columns:
    if DF_COPY[col].dtype == "int64":
        value_counts = DF_COPY[col].value_counts()
        print(f"Value counts for the column '{col}' are: \n\n{value_counts}\n")
        
        # Create a figure with three subplots: one for displot, one for scatter plot, and one for boxplot
        fig, ax = plt.subplots(3, 1, figsize=(5, 5), gridspec_kw={'height_ratios': [5, 5, 2]})
        
        # Plotting the histogram and KDE
        sns.histplot(DF_COPY[col], kde=True, ax=ax[0])
        ax[0].set_title(f'Distribution of {col}')
        
        # Plotting the scatter plot
        sns.scatterplot(x=DF_COPY[col], y=DF_COPY.index, ax=ax[1])
        ax[1].set_title(f'Scatter Plot of {col}')
        ax[1].set_xlabel(col)
        ax[1].set_ylabel('Index')
        
        # Plotting the boxplot
        sns.boxplot(x=DF_COPY[col], ax=ax[2])
        ax[2].set_title(f'Boxplot of {col}')
        ax[2].set_xlabel(col)
        
        # Display the plots
        plt.tight_layout()
        plt.show()
```

```python
# Select the two columns for the bar graph
Region = DF_COPY["Region"]
Number_of_Cases = DF_COPY["Number of Cases"]

# Create the bar graph
plt.figure(figsize=(5, 5))
plt.bar(Region, Number_of_Cases, color='blue')

# Add titles and labels
plt.title('Bar Graph between Region and Number of Cases')
plt.xlabel('Region')
plt.ylabel('Number of Cases')

# Display the plot
plt.show()
```
