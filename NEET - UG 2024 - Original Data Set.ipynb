```python
# Create a deep copy of the DataFrame
DF_copy = df.copy(deep=True)

DF_copy.head()
```

```python
DF_copy.tail()
```

```python
DF_copy.drop(columns=["dummy_srlno"],inplace=True)

# Comment
## We are dropping the 'unique_column' from the dataset because it contains approximately
### 2.3 million different values. Due to its high uniqueness, this column is likely not 
#### significant for our analysis.
DF_copy.head()
```

```python
for col in DF_copy.columns:
    if DF_copy[col].dtype == "int64":
        value_counts = DF_copy[col].value_counts()
        print(f"Value counts for the column '{col}' are: \n\n{value_counts}\n")
```

```python
for col in DF_copy.columns:
    if DF_copy[col].dtype == "object":
        value_counts = DF_copy[col].value_counts()
        print(f"Value counts for the column '{col}' are: \n\n{value_counts}\n")
```

```python
# Step 1: Identify the column with unique values
column_name = 'state'

# Step 2: Extract unique values
unique_values = DF_copy[column_name].unique()

# Step 3: Create subsets
subsets = {}
for value in unique_values:
    subsets[value] = DF_copy[DF_copy[column_name] == value]

# Now `subsets` is a dictionary where each key is a unique value from the column,
# and the value is the corresponding subset of the DataFrame.

# Access and print the subsets for each unique category
for value in unique_values:
    print(f"Subset for category '{value}':")
    print(subsets[value])
    print()
```

```python
# Optionally, print a specific subset if needed
# Access the subset for category 'A'
subset_A = subsets["ANDAMAN AND NICOBAR ISLANDS (UT)"]
print("Subset for category 'ANDAMAN AND NICOBAR ISLANDS (UT)':")
print(subset_A)
```
