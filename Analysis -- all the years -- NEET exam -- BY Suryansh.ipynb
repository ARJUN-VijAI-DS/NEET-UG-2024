<h2>Report on the Compromise of Exam Sanctity in 2024</h2>
    
<h3>Introduction</h3>
This report examines the data from NEET exams from 2016 to 2024, with a particular focus on identifying irregularities in the year 2024 that suggest a compromise in the sanctity of the exam.

<b>Analysis Overview</b>
The analysis involved examining various metrics such as the number of scam cases, average marks, topper marks, number of toppers, number of successful candidates, and the number of perfect scores. Additionally, ratios like the Successful to Topper Ratio and Successful to Perfect Score Ratio were calculated to provide deeper insights.

<b>Key Findings</b>
    
1.Increase in Scam Cases:

The year 2024 saw a significant increase in scam cases, particularly paper leaks in the North and West regions, with centers 'A' and 'D' being affected. This is a stark contrast to the previous years where no scam cases were reported.

2.Higher Average Marks:

The average marks in 2024 increased to 83.75, compared to the historical average of 77.88. This sudden rise of 6 marks in average marks indicates potential irregularities in the exam process or presence of malpractices while the conduction of exam was going on.

3.Unusually High Topper Marks:

The topper marks in 2024 were higher (99.25) with low variation(in all the regions the topper marks was around 99 marks) compared to previous years (96.69) comparatively high variation , i.e. at some centres the average marks of topper even dropped to 94 to 95 and had even gone higher to around 98 at some centres . This consistency at the high end of the spectrum in topper marks in 2024 suggests possible manipulation.

4.Spike in the Number of Toppers:

The number of toppers in 2024 was significantly higher, with an average of 8.75, compared to an average of 3.44 in previous years. This unusual increase of 4.5 in toppers indicates potential irregularities.

5.Higher Number of Successful Candidates:

The number of successful candidates in 2024 was higher, averaging 195, compared to the historical average of 173.12. This indicates a potential compromise in exam difficulty or grading standards.

6.Observation of Perfect Scores:

There were 3.5 perfect scores on average in 2024, while there were no perfect scores recorded in previous years. This sudden appearance of perfect scores raises suspicions about the exam's integrity.

7.Lower Successful to Topper Ratio:

The Successful to Topper Ratio in 2024 was significantly lower (24.61) as compared to 53.92 , indicating a disproportionately high number of toppers relative to successful candidates, suggesting potential exam manipulation.
    
8.High Successful to Perfect Score Ratio:

The Successful to Perfect Score Ratio was high (89.79) in 2024, indicating an unusual number of successful candidates achieving perfect scores, further suggesting irregularities.

    
<b>Conclusion</b>

The data analysis indicates that the data recorded in 2024 was highly anamolous as compared to the previous years . The presence of several anomalies in the year 2024 that collectively suggest a compromise in the sanctity of the NEET exam. The significant increase in scam cases, higher average marks, unusually high topper marks, spike in the number of toppers, higher number of successful candidates, and the sudden appearance of perfect scores all point towards potential exam irregularities. These findings underscore the need for further investigation into the NEET 2024 exam to ensure the integrity of the examination process.
